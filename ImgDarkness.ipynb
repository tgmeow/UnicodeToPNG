```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
from PIL import Image
from os import listdir
from os.path import isfile, join
```

```python
# IMAGES_DIR = "chars_14-29x29"
IMAGES_DIR = "chars_20-42x42"
IMAGES = [f for f in listdir(IMAGES_DIR) if isfile(join(IMAGES_DIR, f))]
```

```python
def sumLightness(image):
    return sum(sum(sum(image)[:,0:3]))
def readCreateTuple(file_path):
    tmp = imread(join(IMAGES_DIR, file_path))
    lightness = sumLightness(tmp)
    return (lightness, file_path, tmp)
```

```python
read_images = sorted(list(map(readCreateTuple, IMAGES)), key=lambda x: x[0])
```

```python
showcount = 100
for x in range(showcount):
#     plt.subplot(1, showcount, x+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(read_images[x][2])
#     plt.subplots_adjust(wspace=0.5)
    plt.show()
    print("name: {0}, lightness: {1}".format(read_images[x][1],
                                             read_images[x][0]))
```
